```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

df = pd.read_csv(r"C:\Users\ISHITA SAXENA\Desktop\FDS Project\Dataset.csv")
print("\nFirst 5 rows:")
print(df.head())
```

```python
print("\nMissing Value Count")
print(df.isnull().sum())
```

```python
df_cleaned = df.dropna(subset=["Title", "Ingredients"])
print(f"\nOriginal dataset size: {df.shape}")
print(f"Cleaned dataset size: {df_cleaned.shape}")
```

```python
print("Duplicate rows:", df_cleaned.duplicated().sum())
df_cleaned = df_cleaned.drop_duplicates()
print("New dataset size after removing duplicates:", df_cleaned.shape)
```

```python
df_cleaned['Title'] = df_cleaned['Title'].str.title()
print(df_cleaned['Title'].head(10))
```

```python
# Convert all column names to lowercase (just for consistency)
df_cleaned.columns = df_cleaned.columns.str.lower()
# Strip extra spaces from titles
df_cleaned['title'] = df_cleaned['title'].str.strip()

print("Column names after cleaning:", df_cleaned.columns.tolist())
print("\nCleaned Titles:")
print(df_cleaned['title'].head())
```

```python
# Create a new column for number of ingredients
df_cleaned['ingredients_count'] = df_cleaned['ingredients'].apply(lambda x: len(x.split(',')))

# Use IQR to detect outliers
Q1 = df_cleaned['ingredients_count'].quantile(0.25)
Q3 = df_cleaned['ingredients_count'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df_cleaned = df_cleaned[(df_cleaned['ingredients_count'] >= lower_bound) & 
                        (df_cleaned['ingredients_count'] <= upper_bound)]

print("Dataset size after removing ingredient count outliers:", df_cleaned.shape)
```

```python
# Summary statistics for number of ingredients per recipe
print("Summary Statistics for 'ingredients_count':\n")
print(df_cleaned['ingredients_count'].describe())

# Compute extra stats manually
print("\nAdditional Stats:")
print("Mode:", df_cleaned['ingredients_count'].mode()[0])
print("Variance:", df_cleaned['ingredients_count'].var())
print("Standard Deviation:", df_cleaned['ingredients_count'].std())
```

```python
# Count the number of steps by splitting on periods
df_cleaned['steps_count'] = df_cleaned['instructions'] \
    .apply(lambda x: len(str(x).split('.')))

# Plot ingredients_count vs steps_count
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.scatterplot(
    x='ingredients_count',
    y='steps_count',
    data=df_cleaned[:1000]
)
plt.title('Ingredients Count vs Steps Count')
plt.xlabel('Ingredients Count')
plt.ylabel('Steps Count')
plt.show()
```

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

# 1. Load & clean
df = pd.read_csv(r"C:\Users\ISHITA SAXENA\Desktop\FDS Project\Dataset.csv")
df = (
    df
    .dropna(subset=["Title", "Ingredients"])
    .drop_duplicates(subset=["Title", "Ingredients"])
)

df['cleaned_ingredients'] = df['Ingredients'].apply(
    lambda txt: ' '.join(i.strip().lower() for i in re.split(',|;|\n', txt) if i.strip())
)

# 2. Vectorize ingredients
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['cleaned_ingredients'])

# 3. Fit a k-NN model (cosine distance)
nn = NearestNeighbors(n_neighbors=6, metric='cosine', algorithm='brute')
nn.fit(X)

# 4. Recommender class
class RecipeRecommender:
    def __init__(self, df, vectorizer, nn_model):
        self.df = df.reset_index(drop=True)
        self.vect = vectorizer
        self.nn = nn_model

    def recommend(self, user_ingredients, top_n=5):
        cleaned = ' '.join(i.strip().lower() 
                           for i in re.split(',|;|\n', user_ingredients) if i.strip())
        vec = self.vect.transform([cleaned])
        dists, inds = self.nn.kneighbors(vec, n_neighbors=top_n+1)
        recs = self.df.iloc[inds[0][1: top_n+1]]
        return recs[['Title', 'Ingredients']]

# 5. Instantiate
recommender = RecipeRecommender(df, vectorizer, nn)

# 6. Prompt user and display results
if __name__ == "__main__":
    user_input = input("Enter the ingredients you have (comma-separated): ")
    results = recommender.recommend(user_input, top_n=5)
    print(f"\nTop {len(results)} recipe matches for: {user_input}\n")
    for idx, row in results.iterrows():
        print(f"🍽️ {row['Title']}\n   Ingredients: {row['Ingredients']}\n")
```
